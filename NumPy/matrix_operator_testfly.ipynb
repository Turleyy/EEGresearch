```python
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4])
print(a - 2)
print(a*2)
print(a**2)
```

```python
b = np.linspace(0, np.pi*2, 32)
c = np.sin(b)
d = np.cos(b)
plt.plot(b, c, label = 'sin')
plt.plot(b, d, label = 'cos')
plt.legend()
plt.grid(True)
```

```python
a = np.ones((2,3))
b = np.full((3,2), 2)
np.matmul(a, b)
```
